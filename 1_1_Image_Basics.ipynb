```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
```

```python
pic = Image.open('puppy.jpg')
pic
```

```python
type(pic) #this is a jpeg and numpy can't function of it we must convert it to numpy array
```

```python
pic_array = np.array(pic)
```

```python
pic_array.shape #width 1300, height 1950, and 3 color channel
```

```python
plt.imshow(pic_array)
```

```python
#accessing one the channel

pic_red = pic_array.copy()
pic_green= pic_array.copy()
pic_blue = pic_array.copy()

# R G B --> 0 1 2
```

```python
plt.imshow(pic_red[:,:,0]) #the red channel
```

```python
plt.imshow(pic_green[:,:,0], cmap="gray") # make it like one color
```

```python
plt.imshow(pic_green[:,:,0], cmap="Reds") # make it looks like red!
```

```python
plt.imshow(pic_green[:,:,1], cmap="Greens") # the green channel
```

```python
plt.imshow(pic_green[:,:,2], cmap="Blues") # the blue channel
```

```python
#how to see red channel only without matplotlib default coloring function

pic_red[:,:,1]=0 #green channel is zero
pic_red[:,:,2]=0 # blue channel is zero
plt.imshow(pic_red)
```

```python
pic_green[:,:,0]=0 # red channel is zero
pic_green[:,:,2]=0 # blue channel is zero
plt.imshow(pic_green)
```

```python
pic_blue[:,:,0]=0 # red channel is zero
pic_blue[:,:,1]=0 # green channel is zero
plt.imshow(pic_blue)
```
